```python
import matplotlib.pylab as plt
```

```python
#reading the image
im = plt.imread("flowers.jpg")
```

```python
#displaying the image
plt.imshow(im)
```

```python
#getting shape of image to use in for loops
print(im.shape)
```

```python
#applying average method
for x in range(0, 533, 1):
    for y in range(0, 800, 1):
        #(R + G + B) / 3
        newValue=(int(im[x,y,0])+int(im[x,y,1])+int(im[x,y,2]))/3
        im[x,y]=newValue
```

```python
#displaying the image       
plt.imshow(im)
```
