```python
import matplotlib.pylab as plt
```

```python
#reading the image
im = plt.imread("flowers.jpg")
```

```python
#displaying the image
plt.imshow(im)
```

```python
#getting shape of image to use in for loops
print(im.shape)
```

```python
#applying lightness method
for x in range(0, 533, 1):
    for y in range(0, 800, 1):
        #(max(R, G, B) + min(R, G, B)) / 2
        newValue=(int(max(im[x,y,0],im[x,y,1],im[x,y,2]))+int(min(im[x,y,0],im[x,y,1],im[x,y,2])))/2
        im[x,y]=newValue
```

```python
#displaying the image       
plt.imshow(im)
```
